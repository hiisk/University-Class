# 1. 함수
 - 함수란? 하나의 코드집합 묶음으로 반복 작업을 편하게만들고, 코드 재활용성을 높임

## 1-1. 함수의 기본 형태
 - "def 함수명(변수):" 형태로 선언하며 사용할땐 "함수명(값)" 으로 활용 

```
def f(x):
    return 2 * x + 7

def g(x):
    return x ** 2
a = 2
print(f(a) + g(a) + f(g(2)) + g(f(2)))
```

## 1-2. 함수의 유형

```
def a_rectangle_area():         # 매개변수 X , 반환값 X
    print(5 * 7)

def b_rectangle_area(x, y):     # 매개변수 O , 반환값 X
    print(x * y)

def c_rectangle_area():         # 매개변수 X , 반환값 O
    return (5 * 7)
    
def d_rectangle_area(x , y):    # 매개변수 O , 반환값 O
    return (x * y)

a_rectangle_area()
b_rectangle_area(5, 7)
print(c_rectangle_area())
print(d_rectangle_area(5, 7))
```

## 1-3 함수와 변수의 관계

```
def f(x):
    y = x
    x = 5
    return y * y

x = 3
print(f(x))
print(x)
```

```
def spam(eggs):
    eggs.append(1)      # 기존 객체의 주솟값에 1[] 추가
    eggs = [2, 3]       # 새로운 객체 생성

ham = [0]
spam(ham)
print(ham)
```

```
def spam(eggs):
    eggs.append(1)      # 기존 객체의 주솟값에 1[] 추가
    eggs = [2, 3]       # 새로운 객체 생성
    return eggs

ham = [0]
question = spam(ham)
print(ham)
print(question)
```

```
def kwargs_test(**kwargs):
    print(kwargs)
    print("First value is {first}".format(**kwargs))
    print("Second value is {second}".format(**kwargs))
    print("Third value is {third}".format(**kwargs))

kwargs_test(first = 3, second = 4, third = 5)
```

## 1-5 파일 읽고 쓰기

```
f = open("./sample_data/yesterday.txt", 'r')
print(type(f))
```

```
yesterday_lyric = f.readlines()
print(yesterday_lyric)
f.close()

contents = ""
for line in yesterday_lyric:
    contents = contents + line.strip() + "\n"

n_of_yesterday = contents.upper().count("YESTERDAY")
print("Number of a Word 'Yesterday'" , n_of_yesterday)
```

### 딕셔너리 (Dictionary)
   - 키와 아이템으로 구성되어 있다.
   - 순서는 존재하지 않는다 -> 인덱스 존재 X
   - **변수[키] = 아이템**의 형식을 사용한다
   - 키는 해쉬로 저장된다
   

- 초기화 : {}을 사용하고 내부 키와 아이템은 **키 : 아이템**의 형식이다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
b={0:1,
   1:6,
   7:9,
   9:10}

print(a)
print(a['korea'])
print(b[0])
```

```python
c={'0':1,
   '1':6,
   '7':9,
   '9':10}
print(c['0'])
```

#### 딕셔너리 내에서 새로운 항목 추가 & 변경
   - 새로운 키이면 새로운 키, 아이템 쌍을 추가한다
   - 기존의 키이면 기존의 아이템을 변경한다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}

a['korea'] = 'busan'
a['japan'] = 'tyoko'
print(a)
```

- update 함수 : 딕셔너리 끼리의 병합
    - 공통된 키가 존재한다면 파라미터로 넘기는 딕셔너리로 변경된다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
b = {'japan':'tyoko',
     'china':'beijing',
     'korea':'busan'
    }
b.update(a)
print(b)
```

#### 키를 삭제하는 법
   - pop 함수
   - del 함수 사용

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
b = a.pop('korea')
print(b)
print(a)
```

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
del a['korea']

print(a)
```

- clear 함수 : 딕셔너리를 초기화한다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
a.clear()
print(a)
```

#### in 키워드
   - 키가 딕셔너리 내에 존재하는지 True/False으로 반환
   - 실행시간이 상각실행시간(O(1))이다. -> 굉장히 빠르다
   - 참고) 리스트는 O(n)이다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
print('korea' in a)
```

#### 딕셔너리 내에서 아이템(value) 찾기
   - **변수[키]**의 형식으로 value를 접근하다.
       - value가 없는 경우 오류를 일으킨다 (지양)
       - get 함수 사용, Nonee을 반환 (지향)

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}
print(a.get('japan'))
print(a['japan'])
```

### 딕셔너리 내의 모든 키 || 모든 아이템 || 전부 반환하기
   - keys() 함수 사용 -> 키를 반환
   - values() 함수 사용 -> 값을 반환
   - items() 함수 사용 -> 키와 값의 튜플로 반환
   - 그래서 list() 함수를 통해 list로 변환해준다

```python
a = {'korea':'seoul', 
     'canada':'ottaw', 
     'usa':'washington dc'}

print(a.keys())
print(a.values())
print(a.items())

print(list(a.keys()))
print(list(a.values()))
print(list(a.items()))
```

###  집합 (Set)
   - 수학에서의 집합과 동일
   - 딕셔너리에서 키만 활용하는 개념
       - 따라서 초기화를 {}로 사용한다
       - 혹은 set함수를 사용한다 -> 리스트를 집합으로 변환

```python
a = {1, 2, 3, 4, 5, 5, 5, 5, 5}
b = [1, 2, 3, 4, 5, 6, 7, 8, 8, 9, 9, 10]
c = set(b)
print(type(a))
print(type(c))
print(a)
print(c)
```

#### 집합 연산
    - 교집합, 합집합, 차집합, 부분집합 지원한다
    - 새로운 값을 반환한다.

```python
a = {1, 2, 3, 4, 5}
b = {1, 3, 5, 7, 9}
print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(a.issubset(b))
```
